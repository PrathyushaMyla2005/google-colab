```python
#wordnet lemmatizer
import nltk
from nltk.stem import WordNetLemmatizer

# Download the 'wordnet' resource if not already installed
nltk.download('wordnet')
nltk.download('omw-1.4')  # also helps with more languages / mappings

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Example usage
print(lemmatizer.lemmatize("cacti"))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer, SnowballStemmer

# Download necessary resources
nltk.download('wordnet')
nltk.download('omw-1.4')

# Create objects for both lemmatizer and stemmer
lemmatizer = WordNetLemmatizer()
stemmer = SnowballStemmer("english")

# Apply lemmatization and stemming
print(lemmatizer.lemmatize("best"))         # Lemmatization
print(stemmer.stem("sportingly"))           # Stemming
```

```python
from nltk.stem import WordNetLemmatizer, SnowballStemmer

lemmatizer = WordNetLemmatizer()
stemmer = SnowballStemmer("english")

print("Stemming :", stemmer.stem("studies"))      # -> studi
print("Lemmatization:", lemmatizer.lemmatize("studies"))  # -> study
```
